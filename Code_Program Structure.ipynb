```python
# Program Code ::
import datetime
import pandas as pd

class Institute:
    all_student_dataset = list()
    courses = ['python', 'ds', 'ml', 'hadoop', 'aws']
    
    def __init__(self, name, mob, course, is_weekday):
        student_dataset = dict() #Empty dictionary
        student_dataset['s_id'] = self.generate_id()
        student_dataset['name'] = name
        student_dataset['mob'] = mob
        if self.is_course_valid(course):
            student_dataset['course'] = course
        else:
            student_dataset['course'] = 'Course is not supported currently '+course
        student_dataset['is_weekday'] = is_weekday
        self.all_student_dataset.append(student_dataset)   

    def generate_id(self):
        """
        Auto generated id
        """
        return int(datetime.datetime.now().strftime('%S%H%M'))
    
    def is_course_valid(self, name):
        for course in self.courses:
            if course.lower() == name.lower():
                return True
        return False    
    
    def showdata(self):
        print(self.all_student_dataset)
        
    def show_student_data(self, student_id):
        for current_student_dataset in self.all_student_dataset:
            if current_student_dataset['s_id'] == student_id:
                print("Student detail is : ")
                print("Name : ", current_student_dataset['name'])
                print("MOb : ", current_student_dataset['mob'])
                print("Course : ", current_student_dataset['course'])
                print("Is_weekday : ", current_student_dataset['is_weekday'])
                
                
    def show_student_data_in_dataframe(self):
        for current_student in self.all_student_dataset:
            df = pd.DataFrame([current_student])
        print("Show student data using dataframe")
        print(df)
        

def main():
    while True:
        user_input = int(input("enter your choise 1] Add new student 2] Show All Student Data 3] Exit"))
        
        if user_input == 1:
            name = input("Enter your name : ")
            course = input("Enter your course : ")
            mob = input("Enter your mobile number")
            is_weekday = input("If you want weekday batch type YES and for weekend type NO :")
            student = EtlHive(name, course, mob, is_weekday)
        elif user_input == 2:
            student.showdata()
        elif user_input == 3:
            break
        else:
            print("Enter valid input")
        
if __name__ =='__main__':
    main()
```
